```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y=(2*x**2)+(3*x)
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xlim(0,9)
plt.ylim(0,175)
plt.title('My plot')
#plt.savefig('myfirstplot.png')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes_2 = fig.add_axes([0.2,0.5,0.25,0.25])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
axes.plot(x,y);
axes_2.plot(x,y)
```

```python
fig = plt.figure(dpi =500)
axes = fig.add_axes([0,0,1,1])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
axes.plot(x,y);
```

```python
fig = plt.figure(figsize = (2,2))
axes = fig.add_axes([0,0,1,1])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
axes.plot(x,y);
fig.savefig('new_figure.png', bbox_inches = 'tight')
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =2, dpi = 200, figsize =(4,10))
axes[1][1].plot(x,y);

axes[1][0].plot(a,b);
axes[1][0].set_ylabel("Y Label 1,0")

axes[0][0].plot(a,b);
axes[0][0].set_xlabel("X Label 0,0")

axes[0][1].plot(x,y);

fig.suptitle("Figure level", fontsize =16)

fig.tight_layout()

fig.set_figwidth(10)
fig.savefig("new_subplots.png", bbox_inches = "tight")
```
