```python
def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
import time
start_time = time.time()

df2["tip_quality"] = df2[["total_bill", "tip"]].apply(lambda df2: quality(df2["total_bill"], df2["tip"]), axis = 1)
df2

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Elapsed Time: {elapsed_time: .4f} seconds.")
```

```python
import time
start_time = time.time()

df2["tip_quality"] = np.vectorize(quality)(df2["total_bill"], df2["tip"])
df2

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time: .4f} seconds")
```
